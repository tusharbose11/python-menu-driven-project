```python
''' product -  code, name, company, price, stock, sold'''
#creates a new file
def createfile():  
 import pickle
 with open("pr.dat","wb") as f:
  while True:
    code= int(input("Enter code "))
    name=input("Enter name of the product ") 
    company=input("Company:    ")
    price=int(input("Price:  "))
    stock=int(input("Enter stock "))
    sold=0 
    a=[code, name, company, price, stock, sold]
    pickle.dump(a,f)
    ch=input("Add any more(y/n)  ")
    if ch=="n" or ch=="N":
        break

#add new records
def addmoreproducts():  
 import pickle
 with open("pr.dat","ab") as f:
  while True:
    code= int(input("Enter code "))
    name=input("Enter name of the product ") 
    company=input("Company    ")
    price=int(input("Price  "))
    stock=int(input("Enter stock "))
    sold=0
    a=[code, name, company, price, stock, sold]
    pickle.dump(a,f)
    ch=input("Add any more(y/n)  ")
    if ch=="n" or ch=="N":
        break

#print all the info
def printall():
    import pickle
    with open("pr.dat","rb") as f:
        while True:
            try:
                temp=pickle.load(f)
                print("Code      ",temp[0])
                print("Name      ",temp[1])
                print("Company   ",temp[2])
                print("Price     ",temp[3])
                print("Stock     ",temp[4])
                print("Sold      ",temp[5])
                print("@"*40)
            except:
                break
                
#Display acc to product name
def print_productname():
    import pickle
    n=input("Enter product name to be searched  ")
    with open("pr.dat","rb") as f:
        while True:
            try:
              temp=pickle.load(f)
              if temp[1]==n:  
                print("code      ",temp[0])
                print("name      ",temp[1])
                print("company   ",temp[2])
                print("price     ",temp[3])
                print("stock     ",temp[4])
                print("sold      ",temp[5])
                print("@"*40)
            except:
                break
                
#Display acc to code
def print_productcode():
    import pickle
    n=int(input("Enter product code to be searched  "))
    with open("pr.dat","rb") as f:
        while True:
            try:
              temp=pickle.load(f)
              if temp[0]==n:  
                print("code      ",temp[0])
                print("name      ",temp[1])
                print("company   ",temp[2])
                print("price     ",temp[3])
                print("stock     ",temp[4])
                print("sold      ",temp[5])
                print("@"*40)
            except:
                break

#Display acc to company
def print_productcompany():
    import pickle
    n=input("Enter product company to be searched  ")
    with open("pr.dat","rb") as f:
        while True:
            try:
              temp=pickle.load(f)
              if temp[2]==n:  
                print("code      ",temp[0])
                print("name      ",temp[1])
                print("company   ",temp[2])
                print("price     ",temp[3])
                print("stock     ",temp[4])
                print("sold      ",temp[5])
                print("@"*40)
            except:
                break

#Update stock of the existing products
def updatestock():
    c=int(input("Enter code whose stock is to be updated  "))
    import pickle
    records=[]
    with open("pr.dat","rb") as f:
        while True:
            try:
                temp=pickle.load(f)
                if temp[0]==c:
                    s=int(input("Enter stock to be added  " ))
                    temp[4] = temp[4] + s
                    records.append(temp)
                else:
                    records.append(temp)
            except:
                break
            
    with open ("pr.dat","wb") as f:
       for r in records:
           pickle.dump(r,f)
            
#Update price of the product
def updateprice():
    c=int(input("Enter code whose price is to be updated  "))
    import pickle
    records=[]
    with open("pr.dat","rb") as f:
        while True:
            try:
                temp=pickle.load(f)
                if temp[0]==c:
                    pri=int(input("Enter price to be added  " ))
                    temp[3] = temp[3] + pri
                    records.append(temp)
                else:
                    records.append(temp)
            except:
                break
            
    with open ("pr.dat","wb") as f:
       for r in records:
           pickle.dump(r,f)
            
#Delete selected product            
def deleteproduct():
    c=int(input("Enter code whose info to be deleted "))
    import pickle
    records=[]
    with open("pr.dat","rb") as f:
        while True:
            try:
                temp=pickle.load(f)
                if temp[0]!=c:
                    records.append(temp)                   
            except:
                break
            
    with open ("pr.dat","wb") as f:
       for r in records:
           pickle.dump(r,f)

#Reminder for updating the stock when stock reached 3 
def reminder():
    import pickle
    with open("pr.dat","rb") as f:
        print("YOU HAVE TO UPDATE YOUR STOCK OF THE FOLLOWING : ")
        while True:
            try:
                temp=pickle.load(f)
                if temp[4]<3: #WHOSE stock has reached less than 3              
                  print("code      ",temp[0])
                  print("name      ",temp[1])
                  print("company   ",temp[2])
                  print("price     ",temp[3])
                  print("stock     ",temp[4])
                  print("sold      ",temp[5])
                  print("@"*40)
            except:
                break


#Sell and generate bill
def sellproduct():
    c=int(input("Enter code to be sold "))
    import pickle
    records=[]
    with open("pr.dat","rb") as f:
        while True:
            try:
                temp=pickle.load(f)
                if temp[0]==c:
                  q = int(input("How many to be sold for this product "))
                  if temp[4] < q:
                        print("This much quantity is not available")

                  else:
                    gst = temp[3]*(0.18)
                    temp[4]=temp[4]-q  #stock
                    temp[5]=temp[5]+q   #sold
                    name=input("Please enter your Good Name:  ")
                    ph=int(input("Please enter your Phone number: "))
                    print("product code ", temp[0],end="         ")
                    print("product name ",temp[1])
                    print("Company      ",temp[2])
                    print("quantity     ",q)
                    print("You have to pay ",((temp[3]*q)+gst))#include GST calculation
                    records.append(temp)
                else:
                    records.append(temp)
            except:
                break
            
    with open ("pr.dat","wb") as f:
       for r in records:
           pickle.dump(r,f)

#Menu driven program
print("----------------------"+ 
        "Welcome to ABC Cosmetic Company"+ 
        "----------------------")
while True:
    print("\n Menu: \n ")
    print("Press 1 to Create a new file. \n")
    print("Press 2 to Add products to existing records. \n")
    print("Press 3 to Print all the information about products. \n")
    print("Press 4 to Search product according to its name. \n")
    print("Press 5 to Search product according to its code. \n")
    print("Press 6 to Search product according to its company Name. \n")
    print("Press 7 to Update stock of the existing products. \n")
    print("Press 8 to Update price of a product. \n")
    print("Press 9 to Delete certain product. \n")
    print("Press 10 to Send a reminder to update the stock. \n")
    print("Press 11 to Sell the cosmetics and generate bill. \n")
    print("Press 0 to exit the menu. \n")
    
    choice =int(input("Enter the task you want to do: \n"))
    
    if choice == 1:
        createfile()
    elif choice == 2:
        addmoreproducts()
    elif choice == 3:
        printall()
    elif choice == 4:
        print_productname()
    elif choice == 5:
        print_productcode()
    elif choice == 6:
        print_productcompany()
    elif choice == 7:
        updatestock()
    elif choice == 8:
        updateprice()
    elif choice == 9:
        deleteproduct()
    elif choice == 10:
        reminder()
    elif choice == 11:
        sellproduct()
    elif choice == 0:
        break
    else:
        print("Invalid choice, please choose again\n")
        
        
#End of the program, Thank You.
    




    
```
